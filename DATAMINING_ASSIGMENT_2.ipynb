```python
import pandas as pd
```

```python
df = pd.read_csv('/content/employee_performance.csv')
print("Dataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Summary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=10, kde=True, color='blue')
plt.title("Distribution of Monthly_Salary")
plt.xlabel("Monthly_Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Performance_Score'], y=df['Monthly_Salary'], hue=df['Years_of_Experience'])
plt.title("Monthly_Salary vs.Performance_Score")
plt.xlabel("Performance_Score")
plt.ylabel("Monthly_Salary")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
