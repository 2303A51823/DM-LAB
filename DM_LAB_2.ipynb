```python
!pip install seaborn matplotlib pandas numpy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value1': [10, 20, 30, 40, 50],
    'Value2': [15, 25, 35, 45, 55],
    'Value3': [8, 18, 28, 38, 48]
}

df = pd.DataFrame(data)

print("Statistical Summary:")
print(df.describe())

plt.figure(figsize=(8, 5))
plt.plot(df['Category'], df['Value1'], marker='o', label='Value1')
plt.plot(df['Category'], df['Value2'], marker='o', label='Value2')
plt.title('Line Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(df['Value1'], df['Value2'], c='blue', label='Value1 vs Value2')
plt.title('Scatter Plot')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Value1'], color='skyblue', label='Value1')
plt.bar(df['Category'], df['Value2'], bottom=df['Value1'], color='orange', label='Value2')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df['Value1'], bins=5, alpha=0.7, label='Value1')
plt.hist(df['Value2'], bins=5, alpha=0.7, label='Value2')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.pie(df['Value1'], labels=df['Category'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart (Value1)')
plt.show()

corr = df[['Value1', 'Value2', 'Value3']].corr()
plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
